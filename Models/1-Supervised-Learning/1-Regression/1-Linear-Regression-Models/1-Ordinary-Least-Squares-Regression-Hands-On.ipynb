### Setup

```python
# Import required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import math

from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PowerTransformer

from sklearn.model_selection import train_test_split

import statsmodels.api as sm

from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Library Setups
pd.set_option('display.max_rows',500)
pd.set_option('display.max_columns',None)
pd.options.display.float_format = '{:.2f}'.format
sns.set(style='darkgrid', palette='deep', font_scale=1.2, context='paper')
```

```python
# Read the data
df = pd.read_csv("data/cleaned/vehicle_dataset.csv")
df.head()
```

### Data Preporcessing

- Majority of data exploration and preprocessing is done in the notebook `0-Data-Preperation.ipynb`
- Performing additional pre-processing based on requirements below

```python
# Select columns required for the modeling
col_subset = ['Kilometer','Fuel Type','Transmission','Color','Owner','Seller Type','Engine','Max Power','Max Torque',
              'Drivetrain','Length','Width','Height','Seating Capacity','Fuel Tank Capacity','Vehicle_Age','Price']
df_subset = df[col_subset]
df_subset.head()
```

#### Categorical Data

```python
# Categorical columns to be encoded
cat_cols = ['Fuel Type','Transmission','Color','Owner','Seller Type','Drivetrain']
```

***One Hot Encoder***

```python
# Encode categorical columns using one hot encoding
oh_encoder = OneHotEncoder(sparse=False, drop='first', handle_unknown='ignore')

# Fit and transform the categorical data
oh_encoded_data = oh_encoder.fit_transform(df_subset[cat_cols])

# get feature names
oh_encoded_column_names = oh_encoder.get_feature_names_out(cat_cols)

# Create Encoded Dataframe
oh_encoded_dataframe = pd.DataFrame(oh_encoded_data, columns=oh_encoded_column_names)
oh_encoded_dataframe.head()
```

#### Numerical Data

```python
# Numerical columns to be scaled
num_cols = ['Kilometer','Engine','Max Power','Max Torque','Length','Width','Height','Seating Capacity',
            'Fuel Tank Capacity','Vehicle_Age']
```

```python
# Numerical columns to be scaled
num_cols = ['Kilometer','Seating Capacity',
            'Fuel Tank Capacity','Vehicle_Age']
```

***Standard Scaler***

```python
# Use standard scaler to scale down values
ss_scaler = StandardScaler()

# Fit and transform numerical data
ss_scaled_data = ss_scaler.fit_transform(df_subset[num_cols])

# get feature names
ss_scaled_column_names = ss_scaler.get_feature_names_out(num_cols)

# Created scaled dataframe
ss_scaled_dataframe = pd.DataFrame(ss_scaled_data, columns=ss_scaled_column_names)
ss_scaled_dataframe.head()
```

***Power Transformer***

```python
# Use Power Transformer to scale down values
pt_scaler = PowerTransformer()

# Fit and transform numerical data
pt_scaled_data = pt_scaler.fit_transform(df_subset[num_cols])

# get feature names
pt_scaled_column_names = pt_scaler.get_feature_names_out(num_cols)

# Created scaled dataframe
pt_scaled_dataframe = pd.DataFrame(pt_scaled_data, columns=pt_scaled_column_names)
pt_scaled_dataframe.head()
```

#### Combined Scaled and Encoded data

***One Hot + Standard Scaler***

```python
df_processed_v1 = pd.concat([oh_encoded_dataframe,ss_scaled_dataframe], axis=1)
df_processed_v1['Price'] = df_subset['Price']
df_processed_v1.head()
```

```python
# Extract Correlation
corr_matrix_v1 = df_processed_v1.corr()

# Create a heatmap for the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix_v1, annot=False, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Plot")
plt.show()
```

***One Hot + Power Transformer***

```python
df_processed_v2 = pd.concat([oh_encoded_dataframe,pt_scaled_dataframe], axis=1)
df_processed_v2['Price'] = df_subset['Price']
df_processed_v2.head()
```

```python
# Extract Correlation
corr_matrix_v2 = df_processed_v2.corr()

# Create a heatmap for the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix_v2, annot=False, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Plot")
plt.show()
```

### Train Test Split

```python
# OH + SS version
Xtrain_v1, Xtest_v1, ytrain_v1, ytest_v1 = train_test_split(df_processed_v1.iloc[:,:-1], df_processed_v1.iloc[:,-1], 
                                                            test_size=0.2, random_state=42)
Xtrain_v1.shape, Xtest_v1.shape, ytrain_v1.shape, ytest_v1.shape
```

```python
# OH + PT version
Xtrain_v2, Xtest_v2, ytrain_v2, ytest_v2 = train_test_split(df_processed_v2.iloc[:,:-1], df_processed_v2.iloc[:,-1], 
                                                            test_size=0.2, random_state=42)
Xtrain_v2.shape, Xtest_v2.shape, ytrain_v2.shape, ytest_v2.shape
```

### OLS Model

```python
# # Add a constant (intercept) to the feature matrix
# X = sm.add_constant(X)

# Create OLS Model - V1
ols_model_v1 = sm.OLS(ytrain_v1, Xtrain_v1)

# Create OLS Model - V1
ols_model_v2 = sm.OLS(ytrain_v2, Xtrain_v2)
```

### Train and Evaluate

```python
# V1 data Model Fit
results_v1 = ols_model_v1.fit()

# get Summary
results_v1.summary()
```

```python
# V2 data Model Fit
results_v2 = ols_model_v2.fit()

# get Summary
results_v2.summary()
```

### Coefficients

```python
# V1 Get the coefficients
coefficients = results_v1.params

# Print the coefficients
print("Coefficients:")
print(coefficients)
```

```python
# V2 Get the coefficients
coefficients = results_v2.params

# Print the coefficients
print("Coefficients:")
print(coefficients)
```

### Model Predictions and Residuals

```python
# V1 Predictions
predictions_v1 = results_v1.predict(Xtest_v1)

# Calculate residuals
residuals_v1 = ytest_v1 - predictions_v1

# Create scatter plot for predictions
plt.scatter(predictions_v1, ytest_v1, color='blue')
plt.title('Predicted vs. Actual')
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.show()

# Create residual plot
plt.scatter(predictions_v1, residuals_v1, color='red')
plt.axhline(y=0, color='black', linestyle='--')
plt.title('Residual Plot')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.show()

# Calculate RMSE
mse = mean_squared_error(ytest_v1, predictions_v1)
rmse = math.sqrt(mse)
print(f"Root Mean Squared Error: {rmse:.2f}")
```

```python
# v2 Predictions
predictions_v2 = results_v2.predict(Xtest_v2)

# Calculate residuals
residuals_v2 = ytest_v2 - predictions_v2

# Create scatter plot for predictions
plt.scatter(predictions_v2, ytest_v2, color='blue')
plt.title('Predicted vs. Actual')
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.show()

# Create residual plot
plt.scatter(predictions_v2, residuals_v2, color='red')
plt.axhline(y=0, color='black', linestyle='--')
plt.title('Residual Plot')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.show()

# Calculate RMSE
mse = mean_squared_error(ytest_v2, predictions_v2)
rmse = math.sqrt(mse)
print(f"Root Mean Squared Error: {rmse:.2f}")
```

```python
# Library Setups
pd.set_option('display.max_rows',500)
pd.set_option('display.max_columns',None)
pd.options.display.float_format = '{:.2f}'.format
sns.set(style='darkgrid', palette='deep', font_scale=1.2, context='paper')
```

```python
# Read the data
df = pd.read_csv("data/cleaned/vehicle_dataset.csv")
df.head()
```

### Data Preporcessing

```python
# Select columns required for the modeling
col_subset = ['Kilometer','Fuel Type','Transmission','Color','Owner','Seller Type','Engine','Max Power','Max Torque',
              'Drivetrain','Length','Width','Height','Seating Capacity','Fuel Tank Capacity','Vehicle_Age','Price']
df_subset = df[col_subset]
df_subset.head()
```

#### Categorical Data

```python
# Categorical columns to be encoded
cat_cols = ['Fuel Type','Transmission','Color','Owner','Seller Type','Drivetrain']
```

***One Hot Encoder***

```python
# Encode categorical columns using one hot encoding
oh_encoder = OneHotEncoder(sparse=False, drop='first', handle_unknown='ignore')

# Fit and transform the categorical data
oh_encoded_data = oh_encoder.fit_transform(df_subset[cat_cols])

# get feature names
oh_encoded_column_names = oh_encoder.get_feature_names_out(cat_cols)

# Create Encoded Dataframe
oh_encoded_dataframe = pd.DataFrame(oh_encoded_data, columns=oh_encoded_column_names)
oh_encoded_dataframe.head()
```

#### Numerical Data

```python
# Numerical columns to be scaled
num_cols = ['Kilometer','Engine','Max Power','Max Torque','Length','Width','Height','Seating Capacity',
            'Fuel Tank Capacity','Vehicle_Age']
```

***Standard Scaler***

#### Combined Scaled and Encoded data

```python
df_processed = pd.concat([oh_encoded_dataframe,ss_scaled_dataframe], axis=1)
# df_processed = pd.concat([ss_scaled_dataframe], axis=1)
# df_processed = pd.concat([oh_encoded_dataframe], axis=1)
df_processed['Price'] = df_subset['Price']
df_processed.head()
```

```python
# Extract Correlation
corr_matrix = df_processed.corr()

# Create a heatmap for the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=False, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Plot")
plt.show()
```

### Train Test Split

```python
# OH + SS version
Xtrain, Xtest, ytrain, ytest = train_test_split(df_processed.iloc[:,:-1], df_processed.iloc[:,-1], 
                                                            test_size=0.2, random_state=42)
Xtrain.shape, Xtest.shape, ytrain.shape, ytest.shape
```

### OLS Model

```python
# # Add a constant (intercept) to the feature matrix
# X = sm.add_constant(X)

# Create OLS Model
ols_model = LinearRegression()

# Create OLS Model - Ridge Regression
ols_model_RR = Ridge()

# Create OLS Model - Lasso Regression
ols_model_LR = Lasso()
```

### Train and Evaluate

```python
# Define Alphas to explore
hyperparameters = {
    'alpha': list(np.linspace(0,1,num=101))  # List of alpha values to explore
}
```

```python
# OLS Model
ols_model.fit(Xtrain, ytrain)
```

```python
# Ridge Regression

# Perform grid search
grid_search = GridSearchCV(ols_model_RR, hyperparameters, cv=10, scoring='neg_root_mean_squared_error')  # 5-fold cross-validation
grid_search.fit(Xtrain, ytrain)

# Get the best hyperparameters
best_alpha = grid_search.best_params_['alpha']

print(f"Best alpha: {best_alpha}")

# Get the scores for all alphas
alpha_scores = grid_search.cv_results_['mean_test_score']

# # Print the scores for all alphas
# print("Scores for all alphas:")
# for alpha, score in zip(hyperparameters['alpha'], alpha_scores):
#     print(f"Alpha: {alpha}, Mean Squared Error: {-score:.2f}")

# Train a Ridge model with the best alpha
final_ols_model_RR = Ridge(alpha=best_alpha)
final_ols_model_RR.fit(Xtrain, ytrain)
```

```python
# Lasso Regression

# Perform grid search
grid_search = GridSearchCV(ols_model_LR, hyperparameters, cv=10, scoring='neg_root_mean_squared_error')  # 5-fold cross-validation
grid_search.fit(Xtrain, ytrain)

# Get the best hyperparameters
best_alpha = grid_search.best_params_['alpha']

print(f"Best alpha: {best_alpha}")

# Get the scores for all alphas
alpha_scores = grid_search.cv_results_['mean_test_score']

# # Print the scores for all alphas
# print("Scores for all alphas:")
# for alpha, score in zip(hyperparameters['alpha'], alpha_scores):
#     print(f"Alpha: {alpha}, Mean Squared ELRor: {-score:.2f}")

# Train a Ridge model with the best alpha
final_ols_model_LR = Lasso(alpha=best_alpha)
final_ols_model_LR.fit(Xtrain, ytrain)
```

### Coefficients

```python
# Get the coefficients
coefficients = ols_model.coef_
coefficients_RR = final_ols_model_RR.coef_
coefficients_LR = final_ols_model_LR.coef_

# Print the coefficients
print("Coefficients:")
print(coefficients)
print("Coefficients RR:")
print(coefficients_RR)
print("Coefficients LR:")
print(coefficients_LR)
```

### Model Predictions and Residuals

- Performing additional pre-processing based on requirements below

```python
# Select columns required for the modeling
col_subset = ['Kilometer','Fuel Type','Transmission','Color','Owner','Seller Type','Engine','Max Power','Max Torque',
              'Drivetrain','Length','Width','Height','Seating Capacity','Fuel Tank Capacity','Vehicle_Age','Price']
df_subset = df[col_subset]
df_subset.head()
```

#### Categorical Data

```python
# Categorical columns to be encoded
cat_cols = ['Fuel Type','Transmission','Color','Owner','Seller Type','Drivetrain']
```

***One Hot Encoder***

```python
# Encode categorical columns using one hot encoding
oh_encoder = OneHotEncoder(sparse=False, drop='first', handle_unknown='ignore')

# Fit and transform the categorical data
oh_encoded_data = oh_encoder.fit_transform(df_subset[cat_cols])

# get feature names
oh_encoded_column_names = oh_encoder.get_feature_names_out(cat_cols)

# Create Encoded Dataframe
oh_encoded_dataframe = pd.DataFrame(oh_encoded_data, columns=oh_encoded_column_names)
oh_encoded_dataframe.head()
```

#### Numerical Data

```python
# Numerical columns to be scaled
num_cols = ['Kilometer','Engine','Max Power','Max Torque','Length','Width','Height','Seating Capacity',
            'Fuel Tank Capacity','Vehicle_Age']
```

***Standard Scaler***

```python
# Use standard scaler to scale down values
ss_scaler = StandardScaler()

# Fit and transform numerical data
ss_scaled_data = ss_scaler.fit_transform(df_subset[num_cols])

# get feature names
ss_scaled_column_names = ss_scaler.get_feature_names_out(num_cols)

# Created scaled dataframe
ss_scaled_dataframe = pd.DataFrame(ss_scaled_data, columns=ss_scaled_column_names)
ss_scaled_dataframe.head()
```

```python
# Use standard scaler to scale down values - Output
op_ss_scaler = StandardScaler()

# Fit and transform numerical data
op_ss_scaled_data = op_ss_scaler.fit_transform(df_subset[['Price']])

# get feature names
op_ss_scaled_column_names = op_ss_scaler.get_feature_names_out(['Price'])

# Created scaled dataframe
op_ss_scaled_dataframe = pd.DataFrame(op_ss_scaled_data, columns=op_ss_scaled_column_names)
op_ss_scaled_dataframe.head()
```

```python
# OH + SS version
Xtrain, Xtest, ytrain, ytest = train_test_split(df_processed.iloc[:,:-1], df_processed.iloc[:,-1], 
                                                            test_size=0.2, random_state=42)
Xtrain.shape, Xtest.shape, ytrain.shape, ytest.shape
```

```python
# # Add a constant (intercept) to the feature matrix
# X = sm.add_constant(X)

# Create OLS Model
ols_model = LinearRegression()

# Create OLS Model - Ridge Regression
ols_model_RR = Ridge()

# Create OLS Model - Lasso Regression
ols_model_LR = Lasso()
```

```python
# Ridge Regression

# Perform grid search
grid_search = GridSearchCV(ols_model_RR, hyperparameters, cv=10, scoring='neg_root_mean_squared_error')  # 5-fold cross-validation
grid_search.fit(Xtrain, ytrain)

# Get the best hyperparameters
best_alpha = grid_search.best_params_['alpha']

print(f"Best alpha: {best_alpha}")

# Get the scores for all alphas
alpha_scores = grid_search.cv_results_['mean_test_score']

# # Print the scores for all alphas
# print("Scores for all alphas:")
# for alpha, score in zip(hyperparameters['alpha'], alpha_scores):
#     print(f"Alpha: {alpha}, Mean Squared Error: {-score:.2f}")

# Train a Ridge model with the best alpha
final_ols_model_RR = Ridge(alpha=best_alpha)
final_ols_model_RR.fit(Xtrain, ytrain)
```

```python
# Lasso Regression

# Perform grid search
grid_search = GridSearchCV(ols_model_LR, hyperparameters, cv=10, scoring='neg_root_mean_squared_error')  # 5-fold cross-validation
grid_search.fit(Xtrain, ytrain)

# Get the best hyperparameters
best_alpha = grid_search.best_params_['alpha']

print(f"Best alpha: {best_alpha}")

# Get the scores for all alphas
alpha_scores = grid_search.cv_results_['mean_test_score']

# # Print the scores for all alphas
# print("Scores for all alphas:")
# for alpha, score in zip(hyperparameters['alpha'], alpha_scores):
#     print(f"Alpha: {alpha}, Mean Squared ELRor: {-score:.2f}")

# Train a Ridge model with the best alpha
final_ols_model_LR = Lasso(alpha=best_alpha)
final_ols_model_LR.fit(Xtrain, ytrain)
```
