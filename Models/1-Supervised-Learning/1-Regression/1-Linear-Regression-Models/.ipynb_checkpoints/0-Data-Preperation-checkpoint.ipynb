## Data Cleanup

### Engine Cleanup

```python
# Remove CC and make it float
df['Engine'] = df['Engine'].str.lower().str.replace("cc","").astype(float)
```

```python
# Check which Make-Model Combinations do not have engine value
# If some other data points have engine value, we can use that value to fill nulls as same makes have same engine values
make_model_null_engine_df = df[df['Engine'].isna()][['Make','Model']].drop_duplicates()
make_model_null_engine_df
```

```python
# For the make-model combinations above, 
# extract mean value of engine to avoid duplicates and have roughly center value incase different values
mean_engine_df = df.groupby(['Make','Model'])['Engine'].mean().reset_index()
make_model_extracted_engine_df = pd.merge(left=make_model_null_engine_df, right=mean_engine_df, how='left', on=['Make','Model']).dropna()
make_model_extracted_engine_df
```

```python
# The few engine values we could retreive, we will map it back to the main df
make_model_extracted_engine_df = make_model_extracted_engine_df.rename(columns={'Engine':'Engine_New'})
df = pd.merge(left=df, right=make_model_extracted_engine_df, how='left', on=['Make','Model'])
df['Engine'] = np.where(df['Engine'].isna(),df['Engine_New'],df['Engine'])
df = df.drop('Engine_New', axis='columns')
df.isnull().sum()
```

### Max Power Cleanup

```python
# Extract BHP value
df['Max Power'] = df['Max Power'].str.extract('([0-9]+).*').astype(float)
df['Max Power']
```

```python
# Check which Make-Model Combinations do not have BHP value
# If some other data points have BHP value, we can use that value to fill nulls as same makes have same BHP values
make_model_null_bhp_df = df[df['Max Power'].isna()][['Make','Model']].drop_duplicates()
make_model_null_bhp_df
```

```python
# The few bhp values we could retreive, we will map it back to the main df
make_model_extracted_bhp_df = make_model_extracted_bhp_df.rename(columns={'Max Power':'Max Power_New'})
df = pd.merge(left=df, right=make_model_extracted_bhp_df, how='left', on=['Make','Model'])
df['Max Power'] = np.where(df['Max Power'].isna(),df['Max Power_New'],df['Max Power'])
df = df.drop('Max Power_New', axis='columns')
df.isnull().sum()
```

### Buckets based on price

```python
# Quantile Distribution
df['Price'].quantile([0.1,0.25,0.5,0.75,0.8,0.9,0.95,0.99,1])
```

```python
# Based on above price histogram and quantile, we can split the make into
# >= 99 Percentile = Very Expensive
# Between 90 and 99 Percentile = Expensive
# Between 75 and 90 Percentile = Upper Mid Range
# Between 50 and 75 Percentile = Mid Range
# < 50 Percentile = Budget

# Define the quantile boundaries
quantiles = [0, 0.5, 0.75, 0.9, 0.99, 1.0]

# Define the labels for each quantile range
labels = ['Budget', 'Mid Range', 'Upper Mid Range', 'Expensive', 'Very Expensive']

# Create a new column 'price_category' based on quantile values
df['price_category'] = pd.qcut(df['Price'], quantiles, labels=labels)
df.head()
```

```python
df['price_category'].value_counts(normalize=True)
```

### Vehicle Age

```python
# Instead of year number, assuming max year is the latest year, subtract years from that to get increasing values for older cars
df['Vehicle_Age'] = df['Year'].max() - df['Year']
df.head()
```

### Fuel Type Category Cap

```python
# Apart from Diesel, Petrol and CNG, other categories contribute to less than 1%, so we can tag them as others
df['Fuel Type'] = np.where(df['Fuel Type'].isin(['Diesel','Petrol','CNG']),df['Fuel Type'],"Others")
df['Fuel Type'].value_counts(normalize=True)
```

### Color Category Cap

```python
# Apart from 'White','Silver','Grey','Blue','Black','Red', other colors contribute to less than 5%, so we can tag them as others
df['Color'] = np.where(df['Color'].isin(['White','Silver','Grey','Blue','Black','Red']),df['Color'],"Others")
df['Color'].value_counts(normalize=True)
```

### Owner Type Category Cap

```python
# Apart from 'First','Second','Third', other Owners contribute to less than 1%, so we can tag them as others
df['Owner'] = np.where(df['Owner'].isin(['First','Second','Third']),df['Owner'],"Others")
df['Owner'].value_counts(normalize=True)
```
